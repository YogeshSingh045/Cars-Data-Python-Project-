```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = pd.read_csv('/content/cars_1720274473613 - Copy.csv') # to read the data
```

```python
cars.head(5) # to 5 rows of the data
```

```python
cars.isnull().sum()  # to see if there is any null valuse
```

```python
cars['Cylinders'].fillna(cars['Cylinders'].mean(), inplace = True)  # to fill null value with mean
```

```python
cars.isnull().sum()
```

```python
cars.shape # to see how many columns and rows does data have
```

```python
cars.rename(columns={'Make':'Brand'},inplace = True) # change column name
```

```python
cars.head(5)
```

```python
brand10 = pd.DataFrame((cars['Brand'].value_counts()).iloc[:10])
brand10
```

```python
plt.figure(figsize=(15,4))
ax = sns.barplot(x = 'Brand' , y = 'count' , data = brand10 , color="r")
plt.title('Top Selling Brand', fontsize = 15)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
plt.figure(figsize=(15 ,5))
ax1 = sns.countplot(x = 'Type' , data = cars , color="m")
plt.title('Top Selling Car Type')
for bars in ax1.containers:
  ax1.bar_label(bars)
```

```python
x = cars['Origin'].value_counts().index
y = cars['Origin'].value_counts().values
```

```python
plt.figure(figsize=(6,6))
plt.pie(y,labels=x,startangle=60,autopct='%0.2f%%')
plt.legend(loc=1)
plt.show()
```

```python
plt.figure(figsize=(12,4))
ax2 = sns.countplot(x = 'Type' , data = cars , hue = 'Origin')
plt.title('Most Selling Car Type And There Origin')
for bars in ax2.containers:
  ax2.bar_label(bars)
```

```python
profit = cars.MSRP - cars.Invoice
profit.head(5)
```

```python
cars.insert(7,'Profit',profit,True)
```

```python
cars.head(5)
```

```python
plt.figure(figsize=(15,4))
ax3 = sns.barplot(x = 'Type' , y = 'Profit' , data = cars , hue = 'Origin')
plt.title('Cars with the Highest Profit & Origin')
for bars in ax3.containers:
  ax3.bar_label(bars)
```

```python
plt.figure(figsize=(8,5))
ax4 = sns.lineplot(x = 'Origin' , y = 'Profit' , data = cars , hue = 'DriveTrain')
```

```python
plt.scatter(x=cars['Horsepower'], y= cars['MPG_Highway'] , color = 'r')
plt.title('Horsepower vs Mileage in Highways')
plt.ylabel('Mileage' , fontsize = 18)
plt.xlabel('Horsepower' , fontsize = 18)
```

```python
plt.scatter(x=cars['EngineSize'],y=cars['Horsepower'],color='r')
plt.title('EngineSize vs HorsePower', fontsize=15)
plt.xlabel('EngineSize',fontsize=10)
plt.ylabel('Horsepower',fontsize=10)
```

```python
# total number of orders from top 10 states

sales_state = cars.groupby(['Brand'], as_index=False)['Profit'].sum().sort_values(by='Profit', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,3)})
sns.barplot(data = sales_state, x = 'Brand',y= 'Profit')
```

```python
plt.figure(figsize=(15,5))
sales = cars.groupby(['Brand'], as_index = False)['Profit'].sum().sort_values(by = 'Profit' , ascending=False).head(10)
ax5 = sns.barplot(data = sales , x = 'Brand' , y = 'Profit' , color='r')
plt.title('Brand with Their Profit' , fontsize= 18)
for bars in ax5.containers:
  ax5.bar_label(bars)
```
